```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectFromModel
```

```python
# Convert categorical columns to numerical using LabelEncoder
label_encoder = LabelEncoder()
df['HomeTeam'] = label_encoder.fit_transform(df['HomeTeam'])
df['AwayTeam'] = label_encoder.fit_transform(df['AwayTeam'])
df['Referee'] = label_encoder.fit_transform(df['Referee'])
df['FTR'] = label_encoder.fit_transform(df['FTR'])  # Target variable
```

```python
print(df['FTR'])
```

```python
# Map target variable: Home Win: 1, Draw: 0, Away Win: 2
#df['FTR'] = df['FTR'].map({'H': 1, 'D': 0, 'A': 2})
```

```python
# Features and target
X = df.drop(columns=['FTR', 'FTHG','Season', 'HTR','HomeTeam_ID', 'AwayTeam_ID','FTAG','FTAG','Referee','B365H','B365D','B365A','PSH','PSD','PSA','WHH','WHD','WHA','PSCH','PSCD','PSCD','PSCA'])
y = df['FTR']
```

```python
# Split the dataset: Train (before 2024) and Test (2024)
X_train = X[df['Season'] < 2024]
y_train = y[df['Season'] < 2024]
X_test = X[df['Season'] == 2024]
y_test = y[df['Season'] == 2024]

# Save the split datasets to CSV files
X_train.to_csv('X_train.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```
