```python
import pandas as pd
```

```python
df =pd.read_csv("Premier_League_Season.csv")
```

```python
# Goal Difference: Create new features for goal difference, both at half-time and full-time:
df['GoalDiff_HT'] = df['HTHG'] - df['HTAG']
df['GoalDiff_FT'] = df['FTHG'] - df['FTAG']

# Total Shots: Combine shots and shots on target for both teams:
df['TotalHomeShots'] = df['HS'] + df['HST']
df['TotalAwayShots'] = df['AS'] + df['AST']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns to include in the box plot (excluding FTR and HTR)
columns_to_plot = [
    'FTHG',  # Full Time Home Goals
    'FTAG',  # Full Time Away Goals
    'HTHG',  # Half Time Home Goals
    'HTAG',  # Half Time Away Goals
    'HS',    # Home Shots
    'AS',    # Away Shots
    'HST',   # Home Shots on Target
    'AST',   # Away Shots on Target
    'HF',    # Home Fouls
    'AF',    # Away Fouls
    'HC',    # Home Corners
    'AC',    # Away Corners
    'HY',    # Home Yellow Cards
    'AY',    # Away Yellow Cards
    'HR',    # Home Red Cards
    'AR'     # Away Red Cards
]

# Filter the DataFrame to include only the specified columns
data_to_plot = df[columns_to_plot]

# Create a horizontal box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_to_plot, orient='h')

# Set plot labels and title
plt.yticks(rotation=0)
plt.title('Horizontal Box Plot of Football Match Metrics')
plt.xlabel('Values')
plt.ylabel('Metrics')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Combine HomeTeam and AwayTeam into a single series and get unique team names
unique_teams = pd.concat([df['HomeTeam'], df['AwayTeam']]).unique()

# Create a mapping of team names to unique IDs
team_id_mapping = {team: idx for idx, team in enumerate(unique_teams)}

# Map the team names to IDs in the DataFrame
df['HomeTeam_ID'] = df['HomeTeam'].map(team_id_mapping)
df['AwayTeam_ID'] = df['AwayTeam'].map(team_id_mapping)

# Display the number of unique teams and a sample of the team ID mapping
print(f"Number of unique teams: {len(unique_teams)}")
print("Sample team ID mapping:", list(team_id_mapping.items())[:10])
```

```python
# Set the 'Season' column as the index to use tha column later on for splitting our dataset
df.set_index('Season', inplace=True)

# Save the DataFrame to a CSV file with the new index
df.to_csv('Premier_League_Season_With_Team_IDs.csv')

# Optional: Check if everything worked correctly by displaying the first few rows
print(df.head())
```

```python
# Filter the DataFrame to include only the specified columns
df_filtered = df[columns_to_plot]

# Compute the correlation matrix
corr_matrix = df_filtered.corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(
    corr_matrix,
    annot=True,          # Add values in each square
    cmap='coolwarm',     # Color map
    fmt='.2f',           # Format for the values
    linewidths=0.5,      # Line width between squares
    vmin=-1, vmax=1      # Color scale limits
)

# Improve the visual aesthetics
plt.title('Correlation Heatmap of Football Metrics', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = df_filtered.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(corr_matrix)

# The correlation matrix of various football metrics reveals several noteworthy patterns in the relationships between match statistics. Full Time Home Goals (FTHG) and Full Time Away Goals (FTAG) exhibit a negative correlation of -0.14, indicating that while there is a slight tendency for more goals scored by one team, it does not strongly influence the other teamâ€™s performance. Half Time Home Goals (HTHG) and Full Time Home Goals (FTHG) show a strong positive correlation (0.70), suggesting that teams which perform well in the first half are likely to continue this performance through the match. Conversely, Half Time Away Goals (HTAG) is moderately correlated with Full Time Away Goals (FTAG) (0.70), indicating a consistent performance trend for away teams as well.

# The number of Home Shots (HS) and Home Shots on Target (HST) are highly correlated (0.69), reflecting that teams with more shots generally have more on-target attempts. Away Shots (AS) and Away Shots on Target (AST) also show a strong correlation (0.69), similar to home metrics, suggesting that away teams follow a similar pattern of shot accuracy.

# Interestingly, the metrics related to fouls, such as Home Fouls (HF) and Away Fouls (AF), exhibit weak correlations with other variables, indicating that fouls have less impact on the outcome compared to other factors. Similarly, Home Corners (HC) and Away Corners (AC) are moderately correlated with goals scored, indicating a possible link between corner opportunities and scoring potential.

# The correlation between Home Yellow Cards (HY) and Away Yellow Cards (AY) is relatively low (0.08), highlighting that disciplinary actions are more team-specific rather than universally affecting match outcomes. The Home Red Cards (HR) and Away Red Cards (AR), on the other hand, show a minimal negative correlation (-0.005), suggesting that the presence of red cards in a game is not strongly related between home and away teams.

# Overall, the correlation analysis underscores the significant impact of shots and goals on match outcomes, while other factors like fouls and disciplinary actions have a more nuanced or weaker effect.
```
